11. Write a program that takes the radius of a sphere(a floating point number) as input and then outputs the sphere's diameter, circumference, surface area and volume.

```python
# Function for sphere
def sphere():
    
#     user input radius
    radius = int(input("Enter radius of sphere: "))
    
#     calulating diameter
    diameter = 2 * radius
    
#     calulating circumference
    circumference = 2 * 3.14 * radius
    
#     calulating surface area
    surface_area = 4 * 3.14 * (radius**2)
    
#     calulating volume
    volume = ((4/3) * 3.14 * (radius**3))
    
#     display outputs
    print("Dimeter of sphere of radius %s is %s" % (radius, diameter))
    print("Circumference of sphere of radius %s is %s" % (radius, circumference))
    print("Surface Area of sphere of radius %s is %s" % (radius, surface_area))
    print("Volume of sphere of radius %s is %s" % (radius, volume))
    
# calling function
sphere()
```

12. Instructions An object's momentum is its mass multiplied by its velocity. Write a program that accepts an object's mass (in kilograms) and velocity (in meters per second) as inputs, and then outputs its momentum.

```python
# Function for calculating momentum
def momentum():
    
#     user input mass
    mass = int(input("Enter mass (in kilograms): "))
    
#     user input velocity
    velocity = int(input("Enter velocity (in meters per second): "))
    
#     calculating momentum
    output = mass*velocity
    
#     display output
    print("Momentum of Object is ",output)
    
# calling function
momentum()
```

13. The kinetic energy of a moving object is given by the formula KE = ½mv2 where m is the object’s mass and v is its velocity.

Modify the program you created in Project 5 so that it prints the object’s kinetic energy as well as its momentum.

```python
# Function for calculating kinetic energy 
def ke():

#     user input mass
    mass = int(input("Enter mass (in kilograms): "))
    
#     user input velocity
    velocity = int(input("Enter velocity (in meters per second): "))
    
#     calculating kinetic energy
    output = 0.5 * mass * (velocity**2)
    
#     display output
    print("Momentum of Object is ",output)
    
# calling function    
ke()
```

14. An employee’s total weekly pay equals the hourly wage multiplied by the total number of regular hours plus any overtime pay. Overtime pay equals the total overtime hours multiplied by 1.5 times the hourly wage. Write a program that takes as inputs the hourly wage, total regular hours, and total overtime hours and displays an employee’s total weekly pay.

```python
# user input hour wage
hourWage= float(input ("What is your hourly wage?: "))

# user input regular hours
regularHours= float(input ("How many regular hours did you work this week?: "))

# user input over time hours
overtimeHours= float(input ("How many overtime hours did you have this week?: "))

# user input over time wage
overtimeWage= (1.5*hourWage)

# calculating total weekly pay
totalWeeklyPay= (hourWage*regularHours)+(overtimeHours*overtimeWage)

# display total weekly pay
print("Your total weekly pay is: ", totalWeeklyPay)
```
